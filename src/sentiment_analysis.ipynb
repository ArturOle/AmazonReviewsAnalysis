## Sentiment Analysis

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import json

# Number of entries being analyzed and printed
MAX_PRINT = 9

# Printing report
PRINT_REPORT = True


def sentiment_vader(sentence: str, threshold: float=0.10) -> tuple:
    # Initialization of pre-trained model
    vader_sentiment = SentimentIntensityAnalyzer()

    # Processing 
    sentiment_dict = vader_sentiment.polarity_scores(sentence)
    compound = sentiment_dict['compound']

    # Labeling
    if sentiment_dict['compound'] >= threshold :
        overall_sentiment = "Positive"
    elif sentiment_dict['compound'] <= - threshold :
        overall_sentiment = "Negative"
    else:
        overall_sentiment = "Neutral"
  
    return compound, overall_sentiment


with open("cell_phones_and_accessories.json", 'r') as file:
    data = json.load(file)


if PRINT_REPORT:
    # Entry counter
    i = 0
    # As a single product can have multiple reviews we need to iterate over both
    for key in data.keys():
        for entry in data[key]:
            if i == MAX_PRINT:
                break
            print('productID: {key}'.format(key=key))
            print('Review text: {}'.format(entry["text"]))
            print('Sentiment of the review comment: {}'.format(sentiment_vader(entry["text"])))
            print('Review summary: {}'.format(entry["summary"]))
            print('Sentiment of the summary: {}'.format(sentiment_vader(entry["summary"])))
            print('Score: {}'.format(entry["score"]))
            print('Helpfulness: {}\n'.format(entry["helpfulness"]))
            i += 1
    
```
